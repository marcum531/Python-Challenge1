```python
#import dependencies
import pandas as pd
import csv
import os
import datetime as date
import re
import importlib
```

```python
PyBoss_df = pd.read_csv(r"C:\Users\marcu\OneDrive\Desktop\Python_Challenge1\Python-Challenge1\PyBoss\Employee_Data.csv")
PyBoss_df
```

```python
PyBoss2_df = PyBoss_df
PyBoss2_df.dropna(inplace = True)
PyBoss2_df[['First Name', 'Last Name']] = PyBoss2_df.Name.str.split(" ", expand=True)
PyBoss2_df.drop(columns = ['Name'], inplace = True)
PyBoss2_df
```

```python
#PyBoss2_df['DOB'] = PyBoss2_df.DOB.apply(
    #lambda x: pd.to_datetime(x).strftime('%m/%d/%Y'))
PyBoss2_df['DOB'] = pd.to_datetime(PyBoss2_df['DOB'], dayfirst = False, yearfirst = False)
PyBoss2_df
```

```python
PyBoss3_df = PyBoss2_df
PyBoss3_df.SSN = PyBoss2_df.SSN.apply(lambda x: re.sub(r'\d', '*', x, count=5))
PyBoss3_df
```
